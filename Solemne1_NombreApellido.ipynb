# Aplicaremos el modelo de regresion lineal logística antes de realizar la limpieza de datos

```python
x = np.array(df.drop(['target'],1))
y = np.array(df['target'])
x.shape
```

<h4> Observaciones </h4>
Ahora cargamos las variables de las 4 columnas de entrada en X excluyendo la columna “target” con el método drop(). En cambio agregamos la columna “target” en la variable y. Ejecutamos x.shape para comprobar la dimensión de nuestra matriz con datos de entrada de 150 registros por 4 columnas.

```python
y.shape
```

<h4> Observaciones </h4>
Ejecutamos y.shape para comprobar la dimensión de nuestra matriz con datos de entrada de 150 registros por 1 columna.

## Transformación

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=29)
```

```python
#Se escalan todos los datos
escalar = StandardScaler()
x_train = escalar.fit_transform(x_train)
x_test = escalar.transform(x_test)
```

```python
model = linear_model.LogisticRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
print('Score del modelo (train): %.2f' % model.score(x_train, y_train)) 
print('Score del modelo (test): %.2f' % model.score(x_test, y_test)) 
```

<h4> Observaciones </h4>
En base a los registros antes de haber realizado la limpieza correspondiente, confirmamos cuan bueno fue nuestro modelo utilizando model.score() que nos devuelve la precisión media de las predicciones, en nuestro caso del modelo de entramieto  nos dio una precision del 0.97 y del modelo de test un 0.93

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred))
```

<h4> Observaciones </h4>
En este caso vemos que los aciertos fueron del 93.3% pero hay que tener en cuenta que el tamaño de datos era pequeño.


```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, y_pred))
```

# Cálculos del cuartil para las columnas que tiene outliers en este caso son: sepal width (cm)

```python
print("Sepal width (cm)\n")
Q1_sepal_width =df['sepal width (cm)'].quantile(0.25)
print("Primer cuartil",Q1_sepal_width)
Q3_sepal_width = df['sepal width (cm)'].quantile(0.75)
print("Tercer cuartil",Q3_sepal_width)

IQR_sepal_width = Q3_sepal_width-Q1_sepal_width
print("Rango intercuartil",IQR_sepal_width)

Mediana_sepal_width = df['sepal width (cm)'].median()
print("La mediana es :",Mediana_sepal_width)
Valor_minimo_sepal_width = df['sepal width (cm)'].min()
print("Valor minimo :",Valor_minimo_sepal_width)
Valor_maximo_sepal_width = df['sepal width (cm)'].max()
print("Valor maximo es :",Valor_maximo_sepal_width)

#Calculamos los valores de los bigotes inferior y superior
BI_Calculado_sepal_width = (Q1_sepal_width - 1.5 * IQR_sepal_width)
print("BI Calculado:",BI_Calculado_sepal_width)

BS_Calculado_sepal_width = (Q3_sepal_width + 1.5 * IQR_sepal_width)
print("BS Calcular:",BS_Calculado_sepal_width)
```
